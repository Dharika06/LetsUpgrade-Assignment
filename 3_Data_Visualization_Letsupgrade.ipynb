Question 1
Plot using sns.relplot, line plot, box plot on fmri data set


```python
import seaborn as sns #  Python data visualization library based on matplotlib for statistical graphics
import pandas as pd # data analysis and manipulation tool
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fmri = sns.load_dataset("fmri")
fmri
```

## **RELPLOT**

```python
sns.relplot(data = fmri, x = "timepoint", y = "signal")
```

```python
sns.relplot(data = fmri, x = "timepoint", y = "signal", hue = "event")
```

```python
sns.relplot(data = fmri, x = "timepoint", y = "signal", hue = "event", style = "region")
```

```python
sns.relplot(data = fmri, x = "timepoint", y = "signal", hue = "subject", style = "region", col="event")
```

```python
sns.relplot(data = fmri, x = "timepoint", y = "signal", hue = "subject", style = "region", size = "event")
```

# **LINE PLOT**

```python
fmri = sns.load_dataset("fmri")
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri);
```

```python
sns.lineplot(data = fmri, x = "timepoint", y = "signal", hue = "event")
```

```python
sns.lineplot(data = fmri, x = "timepoint", y = "signal", hue = "event", style = "region")
```

```python
sns.lineplot(data = fmri, x = "timepoint", y = "signal", hue = "subject", style = "region", size = "event")
```

# **BOX PLOT**

```python
sns.boxplot(data = fmri, x = "timepoint", y = "signal")
```

```python

sns.boxplot(data = fmri, x = "timepoint", y = "signal", hue = "event")
```

```python
sns.boxplot(data = fmri, x = "timepoint", y = "signal", hue = "region")
```
